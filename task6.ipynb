```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('../TH_Data/100_rows.csv')
df['minmax_norm'] = 1 + (df['4'] - df['4'].min()) / (df['4'].max() - df['4'].min()) * 9
df['mean_norm'] = 1 + (df['4'] - df['4'].mean()) / (df['4'].max() - df['4'].min()) * 9
df.head()
```

```python
fig, graph = plt.subplots(figsize = (15, 8))

graph.hist(df.minmax_norm, bins = 30, alpha = 0.7, edgecolor = 'white', label = 'Min-Max normalization')
graph.hist(df.mean_norm, bins = 30, alpha = 0.7, edgecolor = 'white', label = 'Mean normalization')
graph.set_title('Comparison between Min-Max normalization and Mean normalization')
graph.legend()
graph.set_xlabel('Ratings (normalized 1 - 10)')
```
