```python
def func_get_url_list():        # appropriate to pass element
    url_set = set()
    element2 = browser.find_elements_by_tag_name('a')
    print(len(element2))
    for elem in element2:
        href = elem.get_attribute('href')
        substring  = "https://www.goodreads.com/book/show/"
        if href is not None:
            if 'https://www.goodreads.com/book/show/' in href:
                url_set.add(href)
    return url_set
            
```

```python
l = func_get_url_list()
print(len(l))
```
