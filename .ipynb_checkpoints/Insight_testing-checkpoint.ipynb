```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
from collections import Counter
```

```python
df = pd.read_csv("final_01.csv")
df
```

```python
# Splitting titles into single words
title_df = df['title']

title_words = []

for i in title_df:
    title_words.append(i)

split_title = [i for item in title_words for i in item.split()]

title_dic = Counter(split_title)

sorted_dic = sorted(title_dic.items(), key=lambda x: x[1], reverse=True)

df_title = pd.DataFrame(sorted_dic)

df_title.rename(columns={ 0 : "Words", 1 : "Frequancy"})

#sorted_dic
#df_title


#_______________________________________________________________________________

# for i in title_words:
#     split_title = title_words[0].split()
# sorted_dic = dict(sorted(title_dic.items(), key=lambda item: item[1]))

# for item in sorted_dic:
#     if 
# title_words

# title_words = [i for i in title_words if i.strip()]

# for i in title_words:
#     for j in i:
#         print(j)
# title_words

# title_dic = Counter(title_words)
# title_dic
```
