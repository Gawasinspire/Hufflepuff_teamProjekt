```python
genres_words = []

for i in genres_df:
    genres_words.append(i)

split_genres = [i for item in genres_words for i in item.split()]

genres_dic = Counter(split_genres)

sorted_dic = sorted(genres_dic.items(), key=lambda x: x[1], reverse=True)

df_genres = pd.DataFrame(sorted_dic)


```

```python
# Grabbing only the data we want
df_genres_subset = df_genres
df_genres_subset = df_genres.iloc[1:]
df_genres_subset = df_genres_subset.iloc[:25,:]
df_genres_subset.drop(df_genres_subset.loc[df_genres_subset[0]=='and'].index, inplace=True)
list_unique_genres=df_genres_subset[0].unique()

df_genres_subset

#____________________________________________________________________
#print (df_genres[df_genres["Frequancy"] >= 3])
#df_genres = df_genres[df_genres.Frequancy > 3]
# df_genres['Frequancy'].map(len) > 3
# df_genres = df_genres.drop(df_genres[df_genres.Frequancy < 4].index)
# df_genres
```

```python
# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a rectangular Figure
plt.figure(figsize=(8, 8))

ax_scatter = plt.axes(rect_scatter)
ax_scatter.tick_params(direction='out', length=6, width=2, colors='r',labelrotation=90,grid_color='b', grid_alpha=0.5)
ax_histx = plt.axes(rect_histx)
ax_histx.tick_params(direction='in', labelbottom=False)
ax_histy = plt.axes(rect_histy)
ax_histy.tick_params(direction='in', labelleft=False)


x=df['genres']

# the scatter plot:
ax_scatter.scatter(x, df['avg_rating'])

# now determine nice limits by hand:
binwidth = 0.25

#ax_scatter.set_xlim((1900, 2050))
ax_scatter.set_ylim((0, df['avg_rating'].max()))

binsx = np.arange(1900, 2050 + binwidth, binwidth)
binsy = np.arange(4, df['avg_rating'].max() + binwidth, binwidth)
ax_histx.hist(x, bins=binsx)
ax_histy.hist(df['avg_rating'], bins=binsy, orientation='horizontal')

ax_histx.set_xlim(ax_scatter.get_xlim())
ax_histy.set_ylim(ax_scatter.get_ylim())

plt.show()
```
